**Diccionario de datos**

**orden_compra**: Número o identificador único de la orden de compra.  
**fecha_creacion**: Fecha de emisión de la orden de compra.  
**fecha_inicio**: Fecha estimada de inicio del servicio u obra contratada.  
**proyecto**: Nombre o código del proyecto (e.g. “Proyecto Cobre Solar”).  
**etapa_proyecto**: Etapa principal del proyecto (Engineering, Procurement, Construction, Owner Cost).  
**rubro**: Rubro específico del gasto (e.g. “Ingeniería de Detalle”, “Compra de Equipos Críticos”).  
**detalle_servicio**: Descripción detallada del bien o servicio contratado.  
**proveedor**: Nombre de fantasía del proveedor.  
**tipo_contrato**: Tipo de contrato acordado (Hora Hombre, Lump Sum, Fijo Mensual, Por Hito).  
**duracion_meses**: Duración estimada del servicio en meses.  
**moneda**: Tipo de moneda en la que se expresa el presupuesto (e.g. USD).  
**monto_presupuestado**: Monto original presupuestado para la orden de compra.  
**monto_real**: Monto efectivamente devengado o ejecutado.  
**desvio_monto_usd**: Diferencia absoluta en USD entre monto real y presupuestado.  
**porcentaje_desvio**: Porcentaje de desviación sobre el presupuesto.  
**avance_estimado_pct**: Porcentaje de avance físico o financiero estimado.  
**tipo_gasto_proyecto**: Clasificación del gasto según su naturaleza (Capex, Opex, Costos del Propietario, etc.).  
**subetapa_epc**: Subetapa específica dentro del modelo EPC (e.g. “Ingeniería de Detalle”, “Obras Civiles”).  
**wbs**: Código del componente en la estructura analítica del proyecto (WBS).  
**responsable_sector**: Sector responsable de la gestión de la orden (Ingeniería, Logística, Construcción, etc.).  
**zona_operativa**: Ubicación geográfica o zona donde se ejecuta la actividad (e.g. Planta, Campamento, S/E).

##Objetivo Analitico

1 - Objetivo Analítico

El objetivo de este análisis es explorar e interpretar el comportamiento económico de las órdenes de compra de un proyecto minero de cobre bajo modalidad EPC (Engineering, Procurement and Construction). A través de la visualización de datos y el análisis estadístico, se busca detectar patrones de desvío entre los montos presupuestados y los montos reales, evaluar la influencia del tipo de contrato sobre el desempeño financiero y generar conocimiento útil para la toma de decisiones estratégicas de control de costos.
Este trabajo funcionará como base para futuros modelos predictivos que puedan anticipar riesgos financieros en proyectos industriales a gran escala.

2 - Contexto

El control de costos en proyectos mineros es crítico para asegurar la viabilidad económica, especialmente en contextos inflacionarios, de volatilidad de insumos o con múltiples contratistas externos. Las empresas deben monitorear continuamente la ejecución económica de contratos para evitar desvíos significativos que comprometan el presupuesto general del proyecto.
En este caso, se analiza un proyecto de infraestructura minera ubicado en el norte argentino, que integra ingeniería, procura, construcción y puesta en marcha de instalaciones para extracción de cobre, incluyendo planta solar, planta concentradora, subestaciones eléctricas y campamentos.

3 - Problema/ Contexto

La ejecución económica del proyecto enfrenta desafíos recurrentes: desvíos en los costos reales respecto al presupuesto inicial, dificultades para controlar servicios con contratos variables, y retrasos que impactan el flujo financiero. Estos desajustes pueden tener múltiples causas: estimaciones poco precisas, deficiencias en la gestión de compras, cambios de alcance o proveedores con bajo desempeño.
Por ello, surge la necesidad de analizar visualmente y de forma multivariada los registros históricos de órdenes de compra, para detectar anomalías, entender tendencias y generar alertas tempranas. #PROYECTO FINAL

Preprocesamiento y Selección de Variables

En esta sección vamos a:

*Estandarizar las variables numéricas.

*Codificar las variables categóricas.

*Seleccionar las ~10–15 features más relevantes mediante un método embedded.




```python
from sklearn.model_selection import train_test_split

# 1. Define target y variables de entrada
target    = 'desvio_monto_usd'
drop_cols = ['orden_compra', 'proyecto', 'monto_real']

X = df.drop(columns=drop_cols + [target])
y = df[target]

# 2. Separa en Train/Test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

print("Tamaño Train:", X_train.shape, "| Tamaño Test:", X_test.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestRegressor

# 1) Preprocesador
num_cols = X_train.select_dtypes(include='number').columns.tolist()
cat_cols = X_train.select_dtypes(exclude='number').columns.tolist()
prepro = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_cols)
])

# 2) Pipeline de selección
sel_pipe = Pipeline([
    ('prepro', prepro),
    ('selector', SelectFromModel(
        RandomForestRegressor(n_estimators=200, random_state=42),
        threshold='median'
    ))
])

# 3) Ajuste únicamente con Train
sel_pipe.fit(X_train, y_train)
# — Extraer correctamente los nombres tras el preprocesamiento —

# 1) Obtener array con todos los nombres de las columnas transformadas
#    (StandardScaler no cambia nombre, OneHotEncoder genera nombres con get_feature_names_out)
feature_names = sel_pipe.named_steps['prepro'].get_feature_names_out()

# 2) Obtener la máscara booleana de SelectFromModel
mask = sel_pipe.named_steps['selector'].get_support()

# 3) Aplicar la máscara sobre el array de nombres
selected_feature_names = feature_names[mask]

# 4) Mostrar resultados
print(f"Features seleccionadas ({len(selected_feature_names)}):")
for name in selected_feature_names:
    print("–", name)

```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Pipeline para Decision Tree con selección embedded
pipe_dt = Pipeline([
    ('prepro', prepro),   # tu ColumnTransformer
    ('selector', SelectFromModel(
        RandomForestRegressor(n_estimators=200, random_state=42),
        threshold='median')),
    ('model', DecisionTreeRegressor(random_state=42))
])

# Pipeline para Random Forest con selección embedded
pipe_rf = Pipeline([
    ('prepro', prepro),
    ('selector', SelectFromModel(
        RandomForestRegressor(n_estimators=200, random_state=42),
        threshold='median')),
    ('model', RandomForestRegressor(n_estimators=300, random_state=42))
])
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. Lista donde guardaremos [modelo, MAE, RMSE, R2]
resultados = []

# 2. Iteramos sobre tus pipelines
for name, pipe in pipes.items():
    # Entrenamos y predecimos
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    # Calculamos métricas
    mae  = mean_absolute_error(y_test, y_pred)
    mse  = mean_squared_error(y_test, y_pred)   # MSE
    rmse = np.sqrt(mse)                          # RMSE
    r2   = r2_score(y_test, y_pred)

    # Guardamos en la lista
    resultados.append([name, mae, rmse, r2])

# 3. Creamos el DataFrame de resultados
df_res = pd.DataFrame(resultados, columns=['Modelo','MAE','RMSE','R2'])
df_res
```

```python
import matplotlib.pyplot as plt

# Elige el modelo a graficar
pipe = pipe_rf      # o pipe_dt
y_pred = pipe.predict(X_test)

plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.5)
lims = [y_test.min(), y_test.max()]
plt.plot(lims, lims, 'r--')
plt.xlabel('Real (desvío USD)')
plt.ylabel('Predicción')
plt.title('Real vs Predicción – Random Forest')
plt.show()
```

```python
import pandas as pd

# Extrae importancias del paso 'model' dentro del pipeline
importances = pipe_rf.named_steps['model'].feature_importances_
feat_names  = sel_pipe.named_steps['prepro'].get_feature_names_out()[sel_pipe.named_steps['selector'].get_support()]

# Top 10
imp_df = pd.Series(importances, index=feat_names).nlargest(10).sort_values()
imp_df.plot.barh(figsize=(6,4))
plt.title('Top 10 importancias (RF)')
plt.xlabel('Importancia')
plt.show()
```
